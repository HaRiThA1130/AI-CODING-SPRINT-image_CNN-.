```python
from ultralytics import YOLO
import cv2

# Load YOLOv8 model
model = YOLO("/kaggle/working/runs/detect/train2/weights/best.pt")  # Using the nano model for efficiency

# Run inference on an image
results = model("/kaggle/input/testimage1/WhatsApp Image 2025-03-14 at 16.36.00_d8be4baf.jpg")  # Replace with your image

# Extract the most probable person
highest_confidence = 0
best_box = None

for result in results:
    for box in result.boxes:  # Iterate through detected objects
        class_id = int(box.cls[0])  # Get class ID
        confidence = box.conf[0].item()  # Get confidence score
        if class_id == 0 and confidence > highest_confidence:  # Class 0 is 'person'
            highest_confidence = confidence
            best_box = box.xyxy[0].tolist()  # Get bounding box

# Draw the best bounding box if found
if best_box:
    img = cv2.imread("image.jpg")
    x1, y1, x2, y2 = map(int, best_box)  # Convert coordinates to integers
    cv2.rectangle(img, (x1, y1), (x2, y2), (0, 255, 0), 2)
    cv2.putText(img, f"Person: {highest_confidence:.2f}", (x1, y1 - 10),
                cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)
    
    # Show image
    cv2.imshow("Most Probable Person", img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
else:
    print("No person detected!")
```

```python
from ultralytics import YOLO
import cv2

# Load the YOLOv8 model
model = YOLO("/kaggle/working/runs/detect/train2/weights/best.pt")

# Run inference on the image
results = model("/kaggle/input/testimage1/WhatsApp Image 2025-03-14 at 16.36.00_d8be4baf.jpg")

# Initialize variables for highest confidence person
best_confidence = 0
best_person_name = None

for result in results:
    for box in result.boxes:
        class_id = int(box.cls[0])  # Get class ID
        confidence = box.conf[0].item()  # Get confidence score
        name = result.names[class_id][0]  # Get detected class name (assuming names are mapped)

        if confidence > best_confidence:
            best_confidence = confidence
            best_person_name = name

# Extract only the first name if a person is found
if best_person_name:
    first_name = best_person_name.split('-')[0]  # Extract the first name
    print("Most probable person:", first_name)
else:
    print("No person detected!")
```
